```python
vk=sns.countplot(x ='Survived',data=df)
for bars in vk.containers:
    vk.bar_label(bars)
```

```python
sns.barplot(x='Sex', y='Survived', data=df)
plt.title('Survival Rate based on Gender')
plt.show()
```

```python
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Survival Rate based on Passenger Class')
plt.show()
```

```python
sns.violinplot(x='Survived', y='Age', data=df)
plt.title('Survival Rate based on Age')
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Number of Survivors based on Passenger Class')
plt.show()
```

```python
sns.boxplot(x='Survived', y='Fare', data=df)
plt.title('Survival Distribution based on Fare')
plt.show()
```

```python
sns.pairplot(df[['Survived', 'Age', 'Fare', 'Pclass', 'SibSp', 'Parch']], hue='Survived')
plt.title('Pairwise Scatter Plot')
plt.show()
```

```python
total_passengers = len(df)
total_survived = df['Survived'].sum()

survival_rate = (total_survived / total_passengers) * 100

print(f"Survival Rate for Titanic Dataset: {survival_rate:.2f}%")
```

```python
survived_count = df['Survived'].sum()
not_survived_count = len(df) - survived_count
plt.figure(figsize=(6, 6)) 
data = [survived_count, not_survived_count]
labels = ['Survived', 'Not Survived']
colors = ['green', 'red']
plt.pie(data, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Survival Rate of Titanic Passengers')
plt.axis('equal')  # Equal aspect ratio ensures that the circle is not distorted
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Survival Rate based on Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
plt.figure(figsize=(6, 6))
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Number of Survivors based on Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
plt.figure(figsize=(8, 6))
sns.countplot(x='Cabin', hue='Survived', data=df)
plt.title('Number of Survivors based on Family Size')
plt.xlabel('Cabin')
plt.ylabel('Count')
plt.show()
```
